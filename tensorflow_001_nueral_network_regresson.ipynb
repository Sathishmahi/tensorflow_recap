```python
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('/content/insurance.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
data['smoker']=pd.get_dummies(data['smoker'],drop_first=True)
```

```python



data.head(1)
```

```python
data['region'].nunique()
```

```python
final_data=pd.concat(((pd.get_dummies(data['region'],drop_first=True),data)),axis=1).drop(columns='region')
```

```python

final_data['sex']=pd.get_dummies(final_data['sex'],drop_first=True)
```

```python

final_data.dtypes
```

```python
x=final_data.drop(columns='charges')
y=final_data['charges']
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train
```

```python
x_train.head()
```

model create

```python
len(x.columns)
```

```python
#create a model
model=tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(8,)),
    tf.keras.layers.Dense(activation='leaky_relu',units=24),
    tf.keras.layers.Dense(activation='leaky_relu',units=48),
    tf.keras.layers.Dense(activation='relu',units=96),
    tf.keras.layers.Dense(activation='relu',units=198),

    tf.keras.layers.Dense(1)


])
```

```python
#model  compile
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.002),
              metrics=['mae'])
```

```python
#model fit
histrory=model.fit(x_train,y_train,epochs=500)
```

```python
model.predict(x_test)[1]
```
